#### Matplotlib

```python
import matplotlib.pyplot as plt 
```

```python
#ejemplo de una linea simple
x = [0, 1, 2, 3, 4, 5]
y = [0, 1, 2, 3, 4, 5]

plt.plot(x, y)
plt.scatter(x, y) #con esto cramos el grafico de dispersion para crear los puntos
plt.xlabel("Eje x") #nombre al eje x
plt.ylabel("Eje y") #nombre al eje y
plt.title("graficas scascateterrr")
plt.show()
```

```python
#grafico de barras
categorias = ["A", "B", "C", "D"]
valores = [10, 20, 25, 39]

plt.bar(categorias, valores)
plt.title("grafico de barras")
plt.xlabel("categorias")
plt.ylabel("valores")
plt.show()
```
